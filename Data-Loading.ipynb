# Load Data

In this File we load the raw data and convert it into JSON format.

Whilst the data was said to be in JSON format, when running it through a JSON linter, we noticed that it was not proper JSON due to the use of single quotes instead of double quotes.

To remedy this, we explored options such as replacing the single quotes with double quotes. However this led to issues where the name of a game included an apostrophe, e.g. `Assassin's Creed`. 


## User-items data

The dataset was downloaded from https://cseweb.ucsd.edu/~jmcauley/datasets.html#steam_data file *Version 1: User and Item Data*.

```python
import pandas as pd
import numpy as np
import ast
import json
```

```python
with open("C:\\Users\\pmogh\\Documents\\Python_Workspace\\Final_Project_Dataset\\ProjectData\\Part_1\\users_items.json",encoding="utf-8") as player_item_file:
    lines = player_item_file.readlines()[0:10000]
```

```python
lines[0]
```

```python
len(lines)
```

Total 10000 lines in data , Every Line display Player Data

```python
data_linter = ast.literal_eval(lines[0])
```

```python
data_linter
```

```python
# Merger Line with comma & Build List
merge_data = '[' + ','.join(lines) + ']'
```

```python
data_linter = ast.literal_eval(merge_data)
```

```python
# Export to JSON

with open('data.json', 'w') as j_file:
    json.dump(data_linter, j_file)
```

We now have a `.json` file that we can easily view as a Pandas DataFrame.

```python
data_frame = pd.DataFrame(data_linter)
data_frame.head()
```
